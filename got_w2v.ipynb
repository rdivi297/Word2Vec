```python
from nltk import sent_tokenize #sentence tokenize
from gensim.utils import simple_preprocess
import os
import gensim
path="/content/drive/MyDrive/archive (1)/"
story=[]
files=os.listdir(path)
for filename in os.listdir(path):

    f = open(os.path.join(path,filename),encoding='latin-1')
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        story.append(simple_preprocess(sent)) #basic level preprocessing required
story
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
model= gensim.models.Word2Vec(
    window=10, #contextual words =10 on either side of the centre word
    min_count=2, #considers sentences with atleast 2 words
    )
```

```python
model.build_vocab(story) #identifies the unique words
```

```python
model.train(story,total_examples=model.corpus_count,epochs=model.epochs) #default value of epochs is 5
```

```python
model.wv.most_similar('arya')
```

```python
model.wv.similarity('jon','daenerys')
```

```python
model.wv.similarity('tyrion','jaime')
```

```python
model.wv.doesnt_match(['cersei','tywin','robb','joffrey'])
```
